```python
import pandas as pd
df = pd.read_csv('/content/flights.csv')
```

```python
df.head()
```

```python
passengers = df['passengers']
plt.hist([passengers], bins=50)

plt.xlabel('Passangers')
plt.ylabel('Count')
plt.title('Distribution of Flights')
# plt.xticks([0, 4, 8, 12, 16, 20], ['0-4', '5-8', '9-12', '13-16', '17-20','21-25'])
plt.show()
```

```python
freq_table = df['year'].value_counts().sort_index()

# Display the frequency distribution table
print(freq_table)
```

```python
plt.scatter(df['year'],df['passengers'])
plt.xlabel('Year')
plt.ylabel('Number of passenger per flight')
plt.title('Distribution of Passengers for each flight over years')
```

```python
# Compute the total number of flights for each month
monthly_flights = df.groupby('month')['passengers'].sum()

# Create the pie chart
plt.pie(monthly_flights.values, labels=monthly_flights.index, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Total Number of Flights by Month')

# Show the plot
plt.show()
```

```python
# Group the data by year and compute the total number of passengers for each year
yearly_passengers = df.groupby(df['year'])['passengers'].sum()

# Create the line chart
plt.plot(yearly_passengers.index, yearly_passengers.values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.title('Passenger Traffic by Year')

# Show the plot
plt.show()
```

```python

# Group the data by month and compute the total number of passengers for each month
monthly_passengers = df.groupby(df['month'])['passengers'].sum()

# Reorder the months by their index in the year
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_passengers = monthly_passengers.reindex(month_order)

# Create the bar graph
plt.bar(month_order, monthly_passengers)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total Passengers')
plt.title('Total Passenger Traffic by Month')
plt.xticks(rotation=45, ha='right')
# Show the plot
plt.show()
```

```python
import seaborn as sns


# Create the density plot
sns.kdeplot(df['passengers'], shade=True)

# Add labels and title
plt.xlabel('Passengers')
plt.ylabel('Density')
plt.title('Density Plot of Passengers')

# Show the plot
plt.show()
```

```python
import seaborn as sns



# Pivot the data to create a matrix of passenger traffic by month and year
flight_matrix = df.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(flight_matrix, annot=True, fmt='d')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Traffic by Month and Year')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt


# Create a scatter plot with bubble sizes representing passenger traffic
plt.scatter(df['year'], df['month'], s=df['passengers']*0.1, alpha=0.5)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Traffic by Month and Year')

# Show the plot
plt.show()
```
